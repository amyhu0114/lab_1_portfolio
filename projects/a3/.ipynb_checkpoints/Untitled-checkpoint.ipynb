```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from datetime import datetime
```

```python
df = pd.read_csv('COVID_US_cases.csv')
df = df[df['new_confirmed'] >= 0]
df['date'] = pd.to_datetime(df['date'])
```

```python
df.describe()
```

```python
df["day_of_year"] = df["date"].dt.dayofyear  # Day of the year (1-365)
df["year"] = df["date"].dt.year

df['logged_new_confirmed'] = np.log(df['new_confirmed']+1) 

# Convert time into polar coordinates
theta = 2 * np.pi * df["day_of_year"] / 365  # Angle for each day
r = df["logged_new_confirmed"]  # Use new_confirmed directly as the radius

# Define a color map based on the year
year_colors = {
    2020: 'green', 
    2021: 'orange', 
    2022: 'blue'
}
colors = df["year"].map(year_colors)

# Plot the spiral bar chart
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection="polar")
bars = ax.bar(theta, r, width=0.02, color=colors, alpha=0.6)

# Formatting
ax.set_title("Spiral Visualization of COVID-19 Cases (2020-2022)", fontsize=14, fontweight='bold')

# Adjust the quarterly markers to be evenly spaced around the circle
quarters = ["Jan", "Apr", "Jul", "Oct"]
quarter_angles = np.linspace(0, 2 * np.pi, 5)  # 5 positions (include 0 for "Jan")

# Set the x-tick labels
ax.set_xticks(quarter_angles[:-1])  # Remove the last angle (it's same as the first one)
ax.set_xticklabels(quarters)

# Set the y-tick labels as the new_confirmed values (just a sample)
r_max = np.log(df["new_confirmed"].max() + 1)
yticks = np.linspace(0, r_max, 5)
ax.set_yticks(yticks)
ax.set_yticklabels([f"{int(val)}" for val in yticks])  # Format the labels as integers

plt.show()
```

```python
# Calculate the rolling 7-day average of 'new_confirmed'
df['7_day_avg'] = df['new_confirmed'].rolling(window=7).mean()

# Drop NaN values that result from the rolling mean calculation
df = df.dropna(subset=['7_day_avg'])

# Extract day of the year and year
df["day_of_year"] = df["date"].dt.dayofyear
df["year"] = df["date"].dt.year

# Convert time into polar coordinates
theta = 2 * np.pi * df["day_of_year"] / 365  # Angle for each day
r = df["7_day_avg"]  # Use the 7-day average as the radius

# Define a color map based on the year
year_colors = {
    2020: 'green',
    2021: 'orange',
    2022: 'blue'
}
colors = df["year"].map(year_colors)

# Plot the spiral bar chart
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection="polar")

# Plot the bars for each 7-day average
for i in range(len(df)):
    ax.bar(theta.iloc[i], r.iloc[i], width=0.02, color=colors.iloc[i], alpha=0.6)

# Formatting
ax.set_title("Spiral Visualization of 7-Day Average COVID-19 Cases (2020-2022)", fontsize=14, fontweight='bold')

# Adjust the quarterly markers to be evenly spaced around the circle
quarters = ["Jan", "Apr", "Jul", "Oct"]
quarter_angles = np.linspace(0, 2 * np.pi, 5)  # 5 positions (include 0 for "Jan")

# Set the x-tick labels
ax.set_xticks(quarter_angles[:-1])  # Remove the last angle (it's same as the first one)
ax.set_xticklabels(quarters)

# Set the y-tick labels as the 7-day average new_confirmed values
r_max = df["7_day_avg"].max()
yticks = np.linspace(0, r_max, 5)
ax.set_yticks(yticks)
ax.set_yticklabels([f"{int(val)}" for val in yticks])  # Format the labels as integers

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming df is your DataFrame with 'date' and 'new_confirmed' columns

# Calculate the rolling 7-day average of 'new_confirmed'
df['7_day_avg'] = df['new_confirmed'].rolling(window=7).mean()

# Drop NaN values that result from the rolling mean calculation
df = df.dropna(subset=['7_day_avg'])

# Extract day of the year and year
df["day_of_year"] = df["date"].dt.dayofyear
df["year"] = df["date"].dt.year

# Define a color map based on the year
year_colors = {
    2020: 'red',
    2021: 'orange',
    2022: 'blue'
}

# Get the maximum value of the 7-day average across all years for consistent scaling
r_max = df['7_day_avg'].max()

# Create a figure with 3 subplots (one for each year)
fig, axes = plt.subplots(1, 3, figsize=(15, 10), subplot_kw={'projection': 'polar'})

# Define months for labels and dotted lines (starting on the 1st day of each month)
month_days = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]  # day_of_year for each month's 1st day
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Loop through each year and plot in the corresponding subplot
for i, year in enumerate([2020, 2021, 2022]):
    ax = axes[i]
    
    # Filter the data for the current year
    year_data = df[df["year"] == year]
    
    # Convert time into polar coordinates
    theta = 2 * np.pi * (year_data["day_of_year"] - 1) / 365  # Angle for each day, subtract 1 to make Jan the top (0 radians)
    r = year_data["7_day_avg"]  # Use the 7-day average as the radius
    
    # Color for the current year
    color = year_colors[year]
    
    # Plot the bars for each 7-day average
    for j in range(len(year_data)):
        ax.bar(theta.iloc[j], r.iloc[j], width=0.02, color=color, alpha=0.6)
    
    # Formatting
    ax.set_title(f"COVID-19 Cases in {year}", fontsize=14, fontweight='bold')
    
    # Adjust the monthly markers to be evenly spaced around the circle (clockwise)
    month_angles = np.linspace(0, 2 * np.pi, 13)  # 12 months, plus one to loop back to Jan
    ax.set_xticks(month_angles[:-1])  # Exclude the last angle, as it's the same as the first one
    ax.set_xticklabels(month_labels, fontsize=16)

    # Set the y-tick labels with increments of 200,000
    yticks = np.arange(0, r_max + 200000, 200000)  # Create y-ticks with step 200,000
    ax.set_yticks(yticks)
    ax.set_yticklabels([f"{int(val)}" for val in yticks])  # Format the labels as integers
    
    # Ensure the same scale for the r-axis across all subplots
    ax.set_ylim(0, r_max)

fig.suptitle('New Covid-19 Cases in United States from January 2020 to September 2022', fontsize=20, fontweight='bold')


# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.cm as cm

# Assuming df is your DataFrame with 'date' and 'new_confirmed' columns

# Calculate the rolling 7-day average of 'new_confirmed'
df['7_day_avg'] = df['new_confirmed'].rolling(window=7).mean()

# Drop NaN values that result from the rolling mean calculation
df = df.dropna(subset=['7_day_avg'])

# Extract day of the year and year
df["day_of_year"] = df["date"].dt.dayofyear
df["year"] = df["date"].dt.year

# Create a figure with a single subplot (polar plot)
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})

# Define months for labels and dotted lines (starting on the 1st day of each month)
month_days = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]  # day_of_year for each month's 1st day
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Define colors for each year
year_colors = {
    2020: 'green',
    2021: 'orange',
    2022: 'blue'
}

# Normalize the 7-day averages to use in colormap for opacity control
norm = plt.Normalize(vmin=df['7_day_avg'].min(), vmax=df['7_day_avg'].max())

# Loop through each year and plot it as a separate line
for year in [2020, 2021, 2022]:
    # Filter the data for the current year
    year_data = df[df["year"] == year]
    
    # Convert time into polar coordinates
    theta = 2 * np.pi * (year_data["day_of_year"] - 1) / 365  # Angle for each day, subtract 1 to make Jan the top (0 radians)
    r = year_data["7_day_avg"]  # Use the 7-day average as the radius
    
    # Get the color for the current year
    color = year_colors[year]
    
    # Adjust the opacity based on the 7-day average values
    opacity = norm(r)  # This gives a value between 0 and 1 for opacity
    
    # Plot the line for this year, with varying opacity based on the 7-day average
    ax.plot(theta, r, color=color, linewidth=2, alpha=0.3)  # Light grey line for the baseline
    ax.scatter(theta, r, color=color, s=5, alpha=opacity)  # Vary the opacity with the 7-day average

# Formatting
ax.set_title('New Covid-19 Cases in United States from January 2020 to September 2022', fontsize=16, fontweight='bold')

# Adjust the monthly markers to be evenly spaced around the circle (clockwise)
month_angles = np.linspace(0, 2 * np.pi, 13)  # 12 months, plus one to loop back to Jan
ax.set_xticks(month_angles[:-1])  # Exclude the last angle, as it's the same as the first one
ax.set_xticklabels(month_labels)

# Set the y-tick labels with increments of 200,000
r_max = df['7_day_avg'].max()
yticks = np.arange(0, r_max + 200000, 200000)  # Create y-ticks with step 200,000
ax.set_yticks(yticks)
ax.set_yticklabels([f"{int(val)}" for val in yticks])  # Format the labels as integers

# Add a colorbar to show the opacity scale
sm = plt.cm.ScalarMappable(cmap=cm.viridis, norm=norm)
sm.set_array([])  # Empty array for the colorbar
plt.colorbar(sm, ax=ax, label="7-day Average New Cases")

# Adjust layout to reduce the space between the title and subplots
plt.tight_layout()

# Show the plot
plt.show()
```
